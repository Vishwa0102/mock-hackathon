```python

import json
#reset
f=open('Student Handout\Input data\level1b.json')
data = json.load(f)
f.close()
```

```python
neighbors=data["neighbourhoods"]
adj={}#adjacency list
g=[]#graph
orderqty=[]
for n,v in neighbors.items() :
    orderqty.append(v["order_quantity"])
    adj[int(n[1:])]=v["distances"]
    x=list(v["distances"])
    g.append(x)
res=data["restaurants"]
rest_dist=res["r0"]["neighbourhood_distance"]
rest_dist.append(0)
adj[50]=rest_dist
g.append(rest_dist)
i=0
for n,dist in adj.items():
    if i<len(rest_dist)-1:
        dist.append(rest_dist[i])
        i+=1
j=0
for node in g:
    if j<len(rest_dist)-1:
        node.append(rest_dist[j])
        j+=1
for k,v in adj.items():
    print(len(v))

vehicle=data["vehicles"]
capacity=vehicle["v0"]["capacity"]
```

```python
start=50
visit=set()
path=[]
allpaths=[]
cost=0

while len(visit)<=50:
    print("satrt = ",start)
    minn=99999
    
    row=g[start][:-1]
    nextt=0
    path.append(start)
    visit.add(start)
    for i in range(len(row)):
        if row[i]<minn:
            if i not in visit and i not in path:
                minn=row[i]
                nextt=i
    if cost+orderqty[nextt]<=capacity:
        start=nextt
        cost+=orderqty[i]
    else:
        cost=0
        start=50
        allpaths.append(path)
        path=[]
print(allpaths)
        
```

```python
inner={}
outer={}
for i in range(len(allpaths)):
    pt=[]
    for j in range(len(allpaths[0])):
        if allpaths[i][j]==50:
            s="r0"
        else:
            s="n"+str(allpaths[i][j])
        

        pt.append(s)
        if s=="n16":
            pt.append("n47")
        if s=="n13":
            pt.append("n9")
    pt.append("r0")
    inner["path"+str(i+1)]=pt

outer["v0"]=inner

json_object = json.dumps(outer)
with open("level1b_output.json", "w") as outfile:
    outfile.write(json_object)
```
