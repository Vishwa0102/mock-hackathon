```python
def findpath(g,start,path,cost):
    if len(path)==21:
        path.append(20)
        return cost
    
    minn=min(g[start])
    l=g[start].copy()
    while l.index(minn) in path:
        l[l.index(minn)]=10000
        minn=min(l)
    nextnode=g[start].index(minn)
    path.append(nextnode)
    cost[0]+=minn
    findpath(g,nextnode,path,cost)
    return cost[0]
```

```python
path=[]
cost=[0]
cost=findpath(g,20,path,cost)
```
