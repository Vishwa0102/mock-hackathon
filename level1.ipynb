```python
def findpath(ratio,start,g,path,cost,capacity,visited):
    #print(visited)
    #print(path)
    #print(cost)
    print("cost ",cost[0])
    print("path ",path)

    minn=min(ratio[start][:-1])
    l=ratio[start][:-1].copy()
    print(l) 
    while l.index(minn) in path or l.index(minn) in visited:
        l[l.index(minn)]=10000
        minn=min(l)
    if cost[0]+orderqty[l.index(minn)]<=capacity:
        nextnode=ratio[start].index(minn)
        path.append(nextnode)
        if nextnode<20:
            cost[0]+=orderqty[nextnode]

        #print("cost=",cost[0])
        findpath(ratio,nextnode,g,path,cost,capacity,visited)
    return cost[0]
```
